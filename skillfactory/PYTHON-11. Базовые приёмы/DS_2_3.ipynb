```python
import pandas as pd
citibike_tripdata = pd.read_csv('data/citibike-tripdata.csv', sep=',')
citibike_tripdata.head()
```

Датасет представляет собой таблицу с информацией о 300 тысячах поездок за первые пять дней сентября 2018 года и включает в себя следующую информацию:

starttime — время начала поездки (дата, время);
stoptime — время окончания поездки (дата, время);
start station id — идентификатор стартовой стоянки;
start station name — название стартовой стоянки;
start station latitude, start station longitude — географическая широта и долгота стартовой стоянки;
end station id — идентификатор конечной стоянки;
end station name — название конечной стоянки;
end station latitude, end station longitude — географическая широта и долгота конечной стоянки;
bikeid — идентификатор велосипеда;
usertype — тип пользователя (Customer — клиент с подпиской на 24 часа или на три дня, Subscriber — подписчик с годовой арендой велосипеда);
birth year — год рождения клиента;
gender — пол клиента (0 — неизвестный, 1 — мужчина, 2 — женщина).

```python
# trip duration starttime	stoptime
# citibike_tripdata['trip duration'] = citibike_tripdata['starttime'] - citibike_tripdata['stoptime']
# display(citibike_tripdata['trip duration'] )
citibike_tripdata_dates = pd.to_datetime(citibike_tripdata['stoptime'], dayfirst = True) - pd.to_datetime(citibike_tripdata['starttime'], dayfirst = True)
display(citibike_tripdata_dates.mean().seconds)
# years_sold = melb_df['Date'].dt.year
citibike_tripdata_dates_seconds = citibike_tripdata_dates.dt.seconds
display(citibike_tripdata_dates_seconds)
citibike_tripdata['trip duration'] = citibike_tripdata_dates_seconds
citibike_tripdata['trip duration'].mean()
```

```python
citibike_tripdata_dayofweek = pd.to_datetime(citibike_tripdata['starttime'], dayfirst = True)
citibike_tripdata_dayofweek = citibike_tripdata_dayofweek.dt.dayofweek
# display(citibike_tripdata_dayofweek)
citibike_tripdata['weekend'] = citibike_tripdata_dayofweek.apply(lambda x: 1 if x in [5,6] else 0)
display(citibike_tripdata['weekend'].value_counts())
```

```python
citibike_tripdata_hourofday = pd.to_datetime(citibike_tripdata['starttime'], dayfirst = True)
citibike_tripdata_hourofday = citibike_tripdata_hourofday.dt.hour
display(citibike_tripdata_hourofday.value_counts())
# display(citibike_tripdata_hourofday)
```

```python
def time_of_day(hour_of_day):
    if hour_of_day in range(0,7):
        return 'night'
    if hour_of_day in range(7,13):
        return 'morning'
    if hour_of_day in range(13,19):
        return 'day'
    if hour_of_day in range(19,24):
        return 'evening'   
print(time_of_day(23))
```

```python
timeofday = citibike_tripdata_hourofday.apply(time_of_day)
# display(timeofday)
citibike_tripdata['time_of_day'] = timeofday
mask_day = citibike_tripdata['time_of_day'] == 'day'
mask_night = citibike_tripdata['time_of_day'] == 'night'
citibike_tripdata[mask_day].shape[0] / citibike_tripdata[mask_night].shape[0]
```
